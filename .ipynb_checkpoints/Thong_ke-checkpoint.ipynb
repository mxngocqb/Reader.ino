```python
# Thông tin tài khoản
MYSQL_ACCOUNT="siot_getdata"
MYSQL_PASSWORD="5eb6d52843c5f3822d6fbb292fafe6ed"
# Lấy dữ liệu trong 1 khoảng thời gian.
DATA_FROM ="2023-01-06 00:00:00"
DATA_TO ="2023-01-08 23:59:00"

DATA_ATTRIBUTE1 = "Id"     # Id của thẻ
DATA_ATTRIBUTE2 = "State"  # Trạng thái của thẻ: Kết nối wifi thành công, kết nối với cổng giai tiếp thành công
```

```python
# tạo đối tượng connection
mydb = mysql.connector.connect(
    host = "siot.soict.ai", 
    user = MYSQL_ACCOUNT, 
    passwd = MYSQL_PASSWORD,
    database = "siot"
)
mycursor = mydb.cursor()
```

```python
# Truy vấn dữ liệu từ thuộc tính State
mycursor.execute("SELECT created_at, value, name FROM AttributesOfDevice "+
                  "where (device_id = '1f8a716f-5a74-489c-81ca-c04b7184ee84')" +
                  "and ((name='{}') or (name='{}')) ".format(DATA_ATTRIBUTE1,DATA_ATTRIBUTE2) +
                  "and (created_at >='{}' and created_at <='{}')".format(DATA_FROM,DATA_TO) +
                  "")

# Lấy dữ liệu
myresult = mycursor.fetchall()
```

```python
# Đưa dữ liệu vào thư viện tính toán 
x1points = np.array([])
x2points = np.array([])
y1points = np.array([])
y2points = np.array([])

for myRecord in myresult:
    if (myRecord[1] == "Hello, you have successfully connected" and myRecord[2] == "State"):
        y1points = np.append(y1points, "Connect")    #  Y:  Hello, you have successfully connected
        x1points = np.append(x1points, myRecord[0])    #  X:  Thời gian ghi lại
    if (myRecord[2] == "Id"):
        y2points = np.append(y2points, myRecord[1])  # Trục y: các thuộc tính
        x2points = np.append(x2points, myRecord[0])  # Trục x: thời gian xuất hiện
```

```python
f, (ax1, ax2) = plt.subplots(2, sharex=True, sharey=False)
plt.xlabel("Datetime")
plt.xticks(rotation=75)         #Xoay nội dung text trục X 90 độ
ax1.plot(x1points, y1points, "o")
ax1.set(title = "Khởi động thiết bị")
ax1.legend(DATA_ATTRIBUTE1)

ax2.plot(x2points, y2points,"o" )
ax2.legend(DATA_ATTRIBUTE2)
ax2.set(title="Trạng thái hoạt động của các thẻ thành viên")
plt.show()
```
